```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

aylar = ['Oca', 'Şub', 'Mar', 'Nis', 'May', 'Haz', 'Tem', 'Ağu', 'Eyl', 'Eki', 'Kas', 'Ara']
satislar = [120, 145, 130, 180, 290, 230, 210, 190, 280, 320, 310, 410]

urunler = ['Cihaz A', 'Cihaz B', 'Cihaz C', 'Cihaz D', 'Diğer']
pazar_payi = [35, 25, 20, 12, 8]

np.random.seed(42)
musteri_yasi = np.random.randint(18, 65, 80)
harcama_miktari = musteri_yasi * 180 + np.random.normal(0, 1200, 80)
teslimat_sureleri = np.random.normal(3, 0.8, 500)

print("Veri Altyapısı Hazır!")
```

```python
fig, ax = plt.subplots(2,2,figsize=(16,10), facecolor='#f4f4f4')
fig.suptitle('2026 Kurumsal Strateji ve Operasyonel Performans Paneli', fontsize=20)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
ax[0, 0].plot(aylar, satislar, color='#1f77b4', marker='o', linewidth=2.5)
ax[0, 0].set_title('Aylık Satış Trendi', fontsize=14, fontweight='semibold')
ax[0, 0].annotate('Yıl Sonu Rekoru!',
                  xy=('Ara', 410),
                  xytext=('Eyl', 380),
                  arrowprops=dict(facecolor='black', shrink=0.05, width=2),
                  fontsize=11,
                  fontweight='bold',
                  color='red')
ax[0, 0].text(0.5, 150, "Kampanya Dönemi", fontsize=10, color='gray', style='italic')
ax[0, 0].grid(True, linestyle='--', alpha=0.5)
```

```python
for row in ax:
    for a in row:
      a.spines['top'].set_visible(False)
      a.spines['right'].set_visible(False)
      a.spines['left'].set_color('#cccccc')
      a.spines['bottom'].set_color('#cccccc')
      a.set_facecolor('white')
      a.title.set_fontsize(14)
      a.title.set_fontweight('bold')
      a.title.set_color('#333333')
      a.tick_params(axis='both', colors='#666666', labelsize=10)
      fig.patch.set_facecolor('#f8f9fa')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 10), facecolor='#f4f4f4')
fig.suptitle('2025 Kurumsal Performans Paneli', fontsize=20)

ax[0, 1].bar(urunler, pazar_payi, color=['#2c3e50', '#34495e', '#7f8c8d', '#95a5a6', '#bdc3c7'])
ax[0, 1].set_title('Ürün Bazlı Pazar Payı (%)')
ax[0, 1].set_ylabel('Pay (%)')

ax[1, 0].scatter(musteri_yasi, harcama_miktari, color='#e67e22', s=80, alpha=0.6, edgecolors='white')
ax[1, 0].set_title('Müşteri Yaşı ve Harcama İlişkisi')
ax[1, 0].set_xlabel('Yaş')
ax[1, 0].set_ylabel('Harcama (TL)')

ax[1, 1].hist(teslimat_sureleri, bins=20, color='#27ae60', edgecolor='white', alpha=0.8)
ax[1, 1].set_title('Teslimat Süresi Dağılımı')
ax[1, 1].set_xlabel('Gün')
ax[1, 1].set_ylabel('Sipariş Sayısı')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Bu panel, birbiriyle kopuk dört grafik yerine, şirketin genel durumunu soldan sağa ve yukarıdan aşağıya bir akışla anlatan bütünsel bir hikâyedir:

1. Trend ve Karşılaştırma (Üst Bölüm: "Durumumuz Ne?")
Sol Üst (Satış Trendi): Hikâyeye zaman serisiyle başlıyoruz. Çizgi grafik kullanıldı çünkü satışlardaki mevsimsel dalgalanmayı ve Aralık ayındaki "Yıl Sonu Rekoru" ivmesini görmenin en etkili yolu budur. Okuyucuya "İşler büyüyor" mesajını ilk saniyede verir.

Sağ Üst (Pazar Payı): Trendin hemen yanına bu başarının mimarlarını koyduk. Sütun grafik (Bar Plot) tercih edildi çünkü kategorik verilerde (Cihaz A, B, C...) kimin lider olduğunu (Cihaz A) ve aradaki farkı net bir şekilde karşılaştırmamıza olanak sağlar.

2. Derin Analiz ve Operasyon (Alt Bölüm: "Neden ve Nasıl?")
Sol Alt (Müşteri İlişkisi): Satışların kaynağına, yani müşteri profiline iniyoruz. Scatter Plot kullanıldı çünkü yaş ve harcama gibi iki sayısal değişken arasındaki korelasyonu (ilişkiyi) anlamak istedik. Grafikteki kümelenme, 40+ yaş grubunun daha yüksek harcama potansiyeline sahip olduğunu (hedef kitlemizi) kanıtlıyor.

Sağ Alt (Teslimat Süreleri): Hikâyeyi operasyonel güvenle bitiriyoruz. Burada Histogram kullandım çünkü sadece ortalama teslimat süresini bilmek yetmez; verinin nasıl dağıldığını (çan eğrisi) görmeliyiz. Çoğu siparişin 3 gün civarında toplanması, operasyonumuzun stabil olduğunu ve büyüme hızına ayak uydurduğunu gösterir.

🟦 Son Hücre – Yönetici Özeti
Büyüme Dinamiği: Aralık ayındaki rekor performans, gelecek yıl için güçlü bir nakit akışı öngörüsü sağlamaktadır.

Ürün Odaklılık: Cihaz A pazarın domine edici gücüdür; ancak Aksesuar/Diğer grubundaki düşük pay için "çapraz satış" (cross-sell) stratejileri geliştirilmelidir.

Müşteri Fırsatı: 40-60 yaş aralığındaki müşterilerin harcama kapasitesi en yüksektir. Bu segmente özel "Sadakat Programları" karlılığı artıracaktır.

Operasyonel Sağlık: Histogramdaki 3 günlük yoğunlaşma, müşteri memnuniyetinin sürdürülebilir olduğunu teyit eder.
